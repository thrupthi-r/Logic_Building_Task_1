```python
#2.Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
passes=0
fail=0
for mark in marks:
    if mark >= 50:
        passes=passes+1
    else:
        fail=fail+1

print(" total number of pass students",passes)
print(" total number of fail students",fail)

```

```python
#3.Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
strlower= []
for name in names:
    strlower.append(name.lower().strip())
    
print("Convert all names to lowercase:",strlower)
```

```python
#4. Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]

for m in messages:
    length = len(m)
    print(f"{m} length is : {length}")

    
    if length >10:
        print(f"'{m}' is longer than 10 characters")
    

    
```

```python
#5.Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
print(logs)
for log in logs:
    if log.lower() == "error":
        count += 1
print("total error count:",count)
    
```
